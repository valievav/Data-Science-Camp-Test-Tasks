**TASK 2**

Consider the matrix of size 20 x 20 filled randomly with binary values 0
(cell is “dead”) or 1 (cell is “alive”).

1) Develop the algorithm that accepts the specified binary array of size
20x20 and iterates the steps where each step executes the following
rules:

● If a living cell has two or three living neighbors, it remains alive;

● if a living cell has one or no living neighbors, it dies of
"loneliness";

● if a living cell has four or more living neighbors, it dies from
"overpopulation";

● if a dead cell has exactly three living neighbors, it comes to life.

2) (Optionally) visualize the process of execution steps.

Response format:
Programming code in jupyter notebook/colab/etc. containing the following:

1) Generating the matrix of size 20x20 with binary random values

2) Algorithm that executes iterations over rules

3) Matrix of values after 20 executed steps

4) (Optionally) visualization of results over iteration process


```python
# Conway’s Game Of Life

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


def display_grid(grid, n=20):
  """
  Display grid with passed values
  """
  fig, ax = plt.subplots()
  img = ax.imshow(grid, cmap ="Greens", interpolation='nearest', extent=[0, n, 0, n])
  ax.grid(True, color="black", linestyle="-", linewidth=0.5)

  # fit grid with cell sizes
  ax.set_xticks(np.arange(n))
  ax.set_yticks(np.arange(n))
  ax.invert_yaxis()  # flip y to respond to correct location of cells
  ax.set_yticklabels(np.arange(n)[::-1])  # adjust y ticks to start from 0

  plt.show()


def get_number_of_neighbors(grid, i, j, verbose=True):
  """
  Calculate number of neighbors for current cell
  (horizontally, vertically, diagonally)
  (i-1, j-1) (i-1, j) (i-1, j+1)
  (i, j-1)   (i, j)   (i, j+1)
  (i+1, j-1) (i+1, j) (i+1, j+1)
  """
  cnt = 0
  for x in range(i-1, i+2):
    for y in range(j-1, j+2):
      # check boundary index is within range, exclude main cell from the check
      if N > x >= 0 and N > y >= 0 and (x != i or y != j) and grid[x, y] == 1:
        cnt += 1

  if cnt != 0 and verbose:
    print(f'Found {cnt} neighbor(s) for cell {i, j}')

  return cnt


def update_grid(grid, verbose=True):
  """
  Apply rules to the grid:
  * alive + < 2 neighbors -> dies of "loneliness"
  * alive + >= 4 neighbors -> dies from "overpopulation"
  * alive + 2-3 neighbors -> stays alive
  * dead + 3 neighbors -> back to life
  """
  new_grid = grid.copy()

  # apply rules to each cell
  for i in range(N):
    for j in range(N):
      neighbors_cnt = get_number_of_neighbors(grid, i, j, verbose)

      # cell is born
      if grid[i, j] == 0 and neighbors_cnt == 3:
        new_grid[i, j] = 1

      # cell dies
      elif grid[i, j] == 1 and (neighbors_cnt < 2 or neighbors_cnt >= 4):
        new_grid[i, j] = 0

  return new_grid
```

```python
# create grid with initial random values
N = 20
grid = np.random.choice([1,0], N*N, p=[0.2, 0.8]).reshape(N, N)
print('Initial grid:')
display_grid(grid)

# run game
runs = 20
for run in range(1, runs+1):
  new_grid = update_grid(grid, verbose=False)
  grid[:] = new_grid[:]
  print(f'Result after {run} run:')
  display_grid(grid)
```
